```python
import matplotlib.pyplot as plt

from focusfusion import focusfuse
from skimage.io import imread
```

# Load and display test images

```python
img1 = imread('clock1.png')[:,:,0]
img2 = imread('clock2.png')[:,:,0]

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.imshow(img1, cmap='gray')
ax1.axis('off')

ax2.imshow(img2, cmap='gray')
ax2.axis('off')

plt.show();
```

# Preview algorithms

```python
algorithms = [
    'guided_filter',
    'sparse_repr',
    'dct_spatial_freq',
    'linear_filters'
]

_, axes = plt.subplots(len(algorithms), 1, figsize=(10, 5 * len(algorithms)))

for i, a in enumerate(algorithms):
    img_fused = focusfuse([img1, img2], algorithm=a)
    
    ax = axes[i]
    ax.imshow(img_fused, cmap='gray')
    ax.set_title(a)
    ax.axis('off')
```
